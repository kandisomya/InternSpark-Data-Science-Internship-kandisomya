```python
df['created Timestamp'] = pd.to_datetime(
    df['created Timestamp'],
    format='%d-%m-%Y %H:%M'
)

df['Hour'] = df['created Timestamp'].dt.hour
df['Day'] = df['created Timestamp'].dt.day_name()

df.head()
```

```python
import matplotlib.pyplot as plt

hourly_comments = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(10,5))
hourly_comments.plot(kind='bar')

plt.title('Comments by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Comments')

plt.show()
```

```python
day_comments = df['Day'].value_counts()

plt.figure(figsize=(8,5))
day_comments.plot(kind='bar')

plt.title('Comments by Day')
plt.xlabel('Day')
plt.ylabel('Number of Comments')

plt.show()
```

```python
emoji_counts = df['emoji used'].value_counts()

plt.figure(figsize=(5,5))
emoji_counts.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Emoji Usage Distribution')
plt.ylabel('')

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['Hashtags used count'].hist(bins=10)

plt.title('Hashtag Usage Distribution')
plt.xlabel('Number of Hashtags')
plt.ylabel('Frequency')

plt.show()
```

```python
top_photos = df['Photo id'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_photos.plot(kind='bar')

plt.title('Top 10 Most Commented Photos')
plt.xlabel('Photo ID')
plt.ylabel('Number of Comments')

plt.show()
```

```python
print("Total Comments:", len(df))
print("Unique Users:", df['User  id'].nunique())
print("Unique Photos:", df['Photo id'].nunique())
print("Average Hashtags Used:", round(df['Hashtags used count'].mean(),2))
```
